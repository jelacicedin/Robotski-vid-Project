```python
import matplotlib.pyplot as plt
%matplotlib notebook
import detObjFunc as RV
orImage=RV.loadImage("petriDish_test_4.png")
RV.showImage(orImage,"original image")
```

```python
def findBacteria(iImage):
    minR=50
    maxR=300
    
    import detObjFunc as RV
    import cv2
    import matplotlib.pyplot as plt
    import detObjFunc as RV
    from PIL import Image
    import numpy as np
    redImg=iImage[:,:,0]
    #grImage=cv2.cvtColor(iImage,cv2.COLOR_RGB2GRAY)
    grImage=cv2.GaussianBlur(redImg, (21,21),0 );
    drImage=iImage
    foundVar=cv2.HoughCircles(grImage,cv2.HOUGH_GRADIENT,1,50,param1=50,param2=50,minRadius=minR,maxRadius=maxR)
    if foundVar is not None:
        foundVar=np.round(foundVar[0,:]).astype("int")
    print(foundVar)
    for (x,y,r) in foundVar:
        drImage=cv2.circle(drImage,(x,y),r,(0,0,255),5)
        
    RV.showImage(drImage,"foundDish")
```

```python
findBacteria(orImage)
```
